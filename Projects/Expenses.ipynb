```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# App title
st.title("üí∞ Personal Finance Tracker Dashboard")

from google.colab import files
uploaded=files.upload()
print(uploaded)

# Assuming 'data' DataFrame is already loaded from a previous cell or 'uploaded' dict.
# If not, you would need to load it here, e.g., data = pd.read_csv(list(uploaded.keys())[0])

data["DATE"] = pd.to_datetime(data["DATE"], format="%d-%m-%Y")

# Total Expense
total_expense = data["AMOUNT"].sum() # Changed to 'AMOUNT' based on kernel state
st.metric("Total Expenses (‚Çπ)", total_expense)

# Category-wise spending
st.subheader("üìä Category-wise Spending")
category_data = data.groupby("CATEGORY")["AMOUNT"].sum() # Changed to 'CATEGORY' and 'AMOUNT'

fig1, ax1 = plt.subplots()
category_data.plot(kind='bar', ax=ax1)
ax1.set_ylabel("Amount (‚Çπ)")
st.pyplot(fig1)

# Pie Chart
st.subheader("üßæ Expense Distribution")
fig2, ax2 = plt.subplots()
ax2.pie(category_data, labels=category_data.index, autopct='%1.1f%%')
ax2.axis('equal')
st.pyplot(fig2)

# Monthly savings prediction
st.subheader("üìà Monthly Savings Prediction")
monthly_income = st.number_input("Enter Monthly Income (‚Çπ)", min_value=0)

if monthly_income > 0:
    savings = monthly_income - total_expense
    st.success(f"Estimated Monthly Savings: ‚Çπ{savings}")
```
